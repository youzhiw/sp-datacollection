```python

df = pd.read_csv("test.txt", header = None, delim_whitespace=True, skiprows = 1, names=[ "network", "station", 
                                                                                        "channel", "two-digit location code", 
                                                                                        "latitude", "longitude", "elevation", 
                                                                                        "phase", "first- motion", "signal onset quality", 
                                                                                        "pick quality", "epicentral distance",  
                                                                                        "time after origin time"])
df
```

```python
def Pphasetime(df, network, station, channel):
    
    nsc = df.loc[(df['station'] == station) & (df['channel'] == channel)& (df['network'] == network)]
    return nsc["time after origin time"][0]
```

```python
time = Pphasetime(df, "CI","CLC","HHZ")
time
```
